Para obtener la información solicitada, el uso de filtros en escencial. A continuacion se presentan una serie de lineas de codigo haciendo uso de funciones de filtos.

FILTRO UNO


*  Eliminar la fila “0” de “Unnamed” y Eliminar la columna “0” y la columna “Referencia”



```python
#ELIMINAR FILA 0
df=pd.read_excel('Inmuebles.xlsx', skiprows=1)
df
```

```python
#Eliminar la columna “Referencia”
df1=df.drop(['Referencia'],axis=1)
df1
```

FILTRO DOS
*   Todos las registros de “Fecha Alta : igual a 2007-04-29 00:00:00 y a 2004-01-02 00:00:00




```python
df1['Fecha_Alta']=df['Fecha_Alta'].astype(str)
df1.info()
```

```python
filtro2_2 = df1[(df1["Fecha_Alta"] == '2007-04-29') | (df1["Fecha_Alta"] == "2004-01-02")]
filtro2_2
```

FILTRO TRES


*   Todos los registros de “ Tipo: parking, local, industrial y piso”



```python
filtro3_2= df1[(df1["Tipo"]== 'Parking' )|(df1["Tipo"]== 'Local')|(df1["Tipo"]== 'Industrial')|(df1["Tipo"]== 'Piso')]
filtro3_2
```

FILTRO CUATRO


*   Todos los registros con “Provincia” igual a “Girona” y con “Superficie” mayor a “100”



```python
filtro4_2 =  df1[(df1["Provincia"]== 'Girona' )&(df1["Superficie"] >100)]
filtro4_2
```

FILTRO CINCO


*   Todos los registros con “Precio Venta: mayor a 1000,000” o “vendedor” igual a “Carmen o Joaquin”



```python
filtro5_2= df1[(df1["Precio Venta"] > 1000000) | (df1["Vendedor"] == "Carmen") | (df1["Vendedor"] == "Joaquín")]
filtro5_2
```

FILTRO SEIS


*   Todos los registros que no sean de “Provincia : Barcelona“



```python
filtro6_2 = df1[~(df1["Provincia"] == "Barcelona")]
filtro6_2
```

FILTRO SIETE


*   Todos los registros que no sean de “Operación: alquiler “ ni “Tipo: oficina“



```python
filtro7_2 = df1[~(df1["Operación"] == "alquiler") & ~(df1["Tipo"] == "oficina")]
filtro7_2
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV | FILTRO1
df1.to_csv("ACT_1.3|FILTRO_1.csv")
filtro2_2.to_csv("ACT_1.3|FILTRO_2.csv")
filtro3_2.to_csv("ACT_1.3|FILTRO_3.csv")
filtro4_2.to_csv("ACT_1.3|FILTRO_4.csv")
filtro5_2.to_csv("ACT_1.3|FILTRO_5.csv")
filtro6_2.to_csv("ACT_1.3|FILTRO_6.csv")
filtro7_2.to_csv("ACT_1.3|FILTRO_7.csv")
```
