Haciendo uso de la funcion .read_excel leemos el archivo que previamente cargamos

```python
#CARGA DESDE UN ARCHIVO .xlsx SIN INDICE
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

Haciendo uso de filtros para variables cualitativas, se hizo uso del filtro por objeto.
Donde se utilizo para buscar todos aquellos de la columna CANT tengan un uno e imprimir todas las filas.



```python
#FILTRO 1 | POR OBJETO
filtro1= df[df['CANT']==1]
filtro1
```

El segundo filtro es por filas. Aqui se busca imprimir las filas consecutivas desde la fila 2 hasta la 22

```python
#FILTRO 2 |  POR FILAS
filtro2= df.iloc[2:22,:] #Filas consecutivas
filtro2

#Imprimir las filas de la 2 hasta la 22
```

```python
#FILTRO 2.1 |  POR FILAS NO CONSECUTIVAS
filtro2_1= df.iloc[[2,22,222],:]
filtro2_1

#Las filas 2, 22, 222; con toda su informacion
```

Este tercer filtro es parecido al anterior, sin embargo este es por colmnnas, en donde busca e imprime las columnas desde la 7 hasta las 13, respectivamente.

```python
#FILTRO 3 |  POR COLUMNAS
filtro3= df.iloc[:,7:13] #Columnas consecutivaas
filtro3

#Todas las columnas consecutivas con datos numericos
```

```python
#FILTRO 3.1 | POR COLUMNAS No CONSECUTIVAS
filtro3_1= df.iloc[:,[2,3,5,6]]
filtro3_1

#Las columnas con datos cualitativas (texto)
```

Para el cuarto filtro, fue necesario indicar el indice, esto se hizo mediante la funcion index_col =. En este caso en particular se indico como indice la columna dos "NOMBRE_VENDEDOR".
Posteriormente, se busco todos aquellos con el nombre de "LETICIA RAMIREZ HERNANDEZ" y que regresara su nombre junto con el nombre_cliente correspondiente.

```python
#FILTRO 4 | INDICAR EL INDICE
df1 = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx',index_col=2)
```

```python
#PASO 2: APLICAR FILRO DE FILAS VS COLUMNAS
filtro4=df1.loc[["LETICIA RAMIREZ HERNANDEZ"], ["NOMBRE_CLIENTE"]]
filtro4
```

El quinto filtro fue por cabecera, este filtro ayuda para observar cierto numeros de registro que se le indique, en este caso fueron los primeros 22.

```python
#FILTRO 5 | DE CABECERA
filtro5=df.head(22)
filtro5

#Los primeros 22 registros
```

El sexto codigo, utiliza un filtro de variables cualitativas.
En este caso, busca de la columna Cantidad todos aquellos que tengan una cantidad mayor a 5, permitiendo observar los comportamientos por fila de todos aquellos que cumplen con esta caracteristica.

```python
#FILTRO 6 |  DE COMPARACION
filtro6= df[df["CANT"] > 5]
filtro6

#Todos aquellos que en la columna de cantidad tengan mas que 5
```

Utilizando un filtro de conector and. Se buscan todos aquellos que de la columan cantidad tengan tres y que en sean publico en general

```python
#FILTRO 7 | Y
filtro7= df[(df["CANT"]== 3)&(df["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL")]
filtro7

# De la columna CANT todos los que tengan 3 y que sean publico en general
```

De igual manera haciando uso de filto de conectores, se utilizo el filtro o, en donde se busca los registros que en el precio unitario tengan menos de 100 o que en la columna de costo unitario sean mayores a 500.

```python
#FILTO 8 | O
filtro8= df[(df["PRECIO_UNITARIO"]< 1000)|(df["COSTO_UNITARIO"] >500)]
filtro8

#Todos Aquellos que en precio unitario tengan menos de 100 o que en costo unitario esten arriba de 500
```

El ultimo filtro hace uso de un filtro de conectores, utilizando el NOT se busca que regrese los registros que no sean LETICIA RAMIREZ HERNANDEZ y que no sean publico en general.

```python
#FILTRO 9 |  NOT
filtro9= df[~(df["NOMBRE_VENDEDOR"]== "LETICIA RAMIREZ HERNANDEZ")&~(df["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL")]
filtro9

#Todos que no sean LETICIA RAMIREZ HERNANDEZ y PÚBLICO EN GENERAL
```
