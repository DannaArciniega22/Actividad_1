```python
#CARGA DESDE UN ARCHIVO .xlsx SIN INDICE
df = pd.read_excel('Datos de Facturación.xlsx')
df
```

```python
#OBTENER INFORMACION DEL DATA FRAME
df.info()
```

Para obtener la información solicitada, el uso de filtros en escencial. A continuacion se presentan una serie de lineas de codigo haciendo uso de funciones de filtos.

```python
#IDENTICAR
df["CVE_CLPV"] = df["CVE_CLPV"].astype(float)
```

```python
#REMPLAZO DE DATOS TIPO STRING A DATOS TIPO NUMERICO
df.CVE_CLPV = df.CVE_CLPV.replace({'O':'0', '6781|': '6781', '3612C':'3612', '      ': '0', '2385A': '2385', '2510A':'2510', '8440Q':'8440', 'M':'0'}, regex=False)
```

```python
#CONVERTILO A FLOAT
df['CVE_CLPV']=df['CVE_CLPV'].astype(float)
```

FILTRO UNO
*   CVE_CLPV (valores: 1000 a 2000)



```python
filtro1_1= df[(df["CVE_CLPV"]>= 1000 )&(df["CVE_CLPV"] <=2000)]
filtro1_1
```

FILTRO DOS
*   CVE_VEND (Todas las claves excepto “5” y “4”)




```python
filtro2_1= df[~(df['CVE_VEND']==5) & ~(df['CVE_VEND']==4)]
filtro2_1
```

```python
#COMPROBAR QUE SE ELIMINARON LOS VALORES
verificar = filtro2_1.groupby(['CVE_VEND'])['CVE_VEND'].count()
verificar
```

FILTO TRES
*   FECHA_ENT   (Fechas de “28/02/2022”)



```python
#CONVERTILO A STR
df['FECHA_ENT']=df['FECHA_ENT'].astype(str)
df['FECHA_ENT']
```

```python
df.info()
```

```python
filtro3_1= df[df['FECHA_ENT']=='2022-04-13 00:00:00']
filtro3_1
```

FILTRO CUATRO
*   CAN_TOT (Cantidades menores a 5951.7)   o  STATUS (“E”)



```python
filtro4_1 = df[(df["CAN_TOT"]< 5951.7)|(df["STATUS"] == "E")]
filtro4_1
```

FILTRO CINCO

*   Solo las columnas: CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT



```python
filtro5_1= df.iloc[ :, [0,6,7,9]]
filtro5_1
```

FILTRO SEIS
*   Solo las filas de 7001-7099



```python
filtro6_1= df.iloc[ 7001: 7099, : ]
filtro6_1
```

FILTRO SIETE
*   Index= CVE_VEND (valores: 1, 2)  (columna: FECHAELAB)



```python
#INDICAR EL INDICE
df1 = pd.read_excel('Datos de Facturación.xlsx',index_col=3)
```

```python
#PASO 2: APLICAR FILRO DE FILAS VS COLUMNAS
filtro7_1=df1.loc[[1,2], ["FECHAELAB"]]
filtro7_1
```
